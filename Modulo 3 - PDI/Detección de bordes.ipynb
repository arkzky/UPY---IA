## Sobel

```python
from skimage import io, filters
from skimage.color import rgb2gray
```

```python
imagen = io.imread("Mariposa.jpg")
io.imshow(imagen)
io.show()
```

```python
imagGris = rgb2gray(imagen)
io.imshow(imagGris)
io.show()
```

```python
sobel = filters.sobel(imagGris)
io.imshow(sobel)
io.show()
```

## Prewit

```python
prewitt = filters.prewitt(imagGris)
io.imshow(prewitt)
io.show()
```
