```python
from skimage import exposure, io, filters
from skimage.color import rgb2gray
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen = io.imread("orquideas.jpg")
io.imshow(imagen)
io.show()
```

```python
imgEq = exposure.equalize_hist(imagen)
plt.imshow(imgEq)
plt.show()
```

```python
imag2 = exposure.equalize_adapthist(imagen, clip_limit = 0.03)
plt.imshow(imag2)
plt.show()
```

```python
io.imsave("act1.jpg",imgEq)
```

```python
io.imsave("act2.jpg", imag2)
```

```python
imag = cv2.imread("orquideas.jpg")
imag1 = cv2.cvtColor(imag, cv2.COLOR_BGR2LAB)
cv2.imshow("image",imag1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
io.imsave("act3.jpg", imag1)
```

```python
img = io.imread("orquideas.jpg")
io.imshow(img)
io.show()
```

```python
imgGris = rgb2gray(img)
io.imshow(imgGris)
io.show()
```

```python
sobel = filters.sobel(imgGris)
io.imshow(sobel)
io.show()
```

```python
io.imsave("act4.jpg", sobel)
```
